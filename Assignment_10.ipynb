1. How do you distinguish between shutil.copy() and shutil.copytree()?

```python
import shutil

copyTree = r"xyz\Python" # give folder path which you want to copy

copyTo = r"xyz\New folder2" # path and mention new folder name where you are going to copy.

shutil.copytree(copyTree, copyTo)

print("successfull")
```

```python
import shutil
copyPath = r"xyz\sample.txt" # give file path which you want to copy

copyTo = r"xyz\new_random.txt" # path, in which file you are going to copy

shutil.copy(copyPath,copyTo)

print("Successfully copied")
```

4.ZipFile objects have a close() method just like File objects’ close() method. What ZipFile method is equivalent to File objects’ open() method?

Ans: open() method itself.

```python
import zipfile

zipFolder = r"D:\FDS\New Text Document.zip"

with zipfile.ZipFile(zipFolder, "r") as archive:
    with archive.open("New Text Document.txt", "r") as file:
        matter = file.read()
        result  = matter.decode().splitlines()
        for resul in result:
            print(resul)
```

5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.

```python
import os
import shutil

sourcePath = "D:\FDS"
newFolder = "D:\FDS\BlueCrystal"
searchfiles = ".jpg"

def copy_files(srcpth,nwfldr,srchfls):
    for x, y, files in os.walk(srcpth):
        for f in files:
            if f.endswith(srchfls):
                source_path = os.path.join(x,f)
                dest_path = os.path.join(nwfldr,f)
                try:
                    shutil.copy(source_path, nwfldr)
                    print(f"copied successfully into {f}")
                except shutil.SameFileError:
                    print(f"Skipped{f} because it found as duplicate.")
                          
copy_files(sourcePath, newFolder, searchfiles)
```
